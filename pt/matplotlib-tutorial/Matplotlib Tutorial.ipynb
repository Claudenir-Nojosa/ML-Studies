![Anatomy of a Matplotlib Figure](https://matplotlib.org/stable/_images/anatomy.png)


```python
# fig means figure, where we're going to put all of your graphs
# ax means axes, kind of the plot itself, the x axis, the y axis
# On the example below we create one figure with one plot, one 'axes'
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
# now we are going to create one figure with two plot, two 'axes
# 1 is the number of rows, 2 is the number of columns
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[1].plot(x, x ** 3)
plt.show()
```

```python
# Create a figure with 4 axes = 2 rows of 2 graphs on each row
fig, axes = plt.subplots(2, 2)
# axes[0].plot(x, y)
# axes[1].plot(x, x ** 3)
plt.show()
```

```python
# Everything inside of a single figure!
# Now let's include the plots of the second column of the first and second row
fig, axes = plt.subplots(2, 2)
axes[0, 1].plot(x, y)
axes[1, 1].plot(x, x ** 3)
plt.show()
```

```python
# But the lines are really clunky, let's make it look smoothier
# The lines are ugly because we only have 7 datapoints, let's see them using another plot called 'scatter'
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

```python
# So let's make more points using arange
x = np.arange(-5, 5, 0.01)
y = x ** 2
print(f"Number of datapoints: {len(x)}")
```

```python
# now that we have more data, the line becomes smoothier, as seen below.
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## Scatterplot, coloring by labels

Let's try coloring the data if the person has Heart Diseases or not, based of the Cholesterol and RestBP

```python
df.head()
```

```python
# Let's change the column AHD for numerical values
df['AHD_Numerical'] = df['AHD'].map({
    'No': 0,
    'Yes': 1
})
```

```python
df['AHD_Numerical'].value_counts()
```

```python
df.head()
```

```python
# Now we create a subset of only the heart diseased person
df_subset_sick = df[df['AHD_Numerical'] == 1]
```

```python
# Then a subset of only the individual who does not have a heart diseased
df_subset_healthy = df[df['AHD_Numerical'] == 0]
```

```python
# Plot the data of all the individuals
fig, ax = plt.subplots()

ax.scatter(df['RestBP'].values, df['Chol'].values)

plt.show()
```

```python
# Plot the data of the individual who does not have heart diseases
fig, ax = plt.subplots()

ax.scatter(df_subset_healthy['RestBP'].values,
           df_subset_healthy['Chol'].values, label='Sick')

# Plot the data of the individual who does have heart diseases

ax.scatter(df_subset_sick['RestBP'].values,
           df_subset_sick['Chol'].values, label='Healthy')

ax.legend()
plt.show()
```

# Section #5 - Using Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# First Graph
axs[0, 0].bar(df_groupedby_age.index, df_groupedby_age['RestBP'].values)
axs[0, 0].set_title('Average Resting Blood Pressure by Age')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Average Resting Blood Pressure')
# Secong Graph
axs[0, 1].plot(df_groupedby_age.index, df_groupedby_age['Chol'].values)
axs[0, 1].set_title('Average Resting Blood Pressure by Age')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Average Resting Blood Pressure')
# Third Graph
axs[1, 0].plot(df_groupedby_age.index, df_groupedby_age['MaxHR'].values)
axs[1, 0].set_title('Average Maximum Heart Rate by Age')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Average Maximum Heart Rate')
# Fourth Graph
axs[1, 1].scatter(df_groupedby_age.index, df_groupedby_age['RestECG'].values)
axs[1, 1].set_title('Average Resting CG by Age')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Average Resting CG')

# Tight Layout makes the plots not overlapping
plt.tight_layout()
plt.show()
```

# Section #6 - Common Questions

## How do we change the figure size?

```python
# Using figsize inside of subplots, in the example below 12 controls the width, the 8 controls the height
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, y)

plt.show()
```

## How do we save our image to a file?

```python
# Basically all you have to do is call 'savefig'
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, y)

plt.savefig('my-plot.png')
```

## How do we change the font size? (Title, axis labels, and tick labels.)

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, y)
ax.set_title('X by Y', fontsize=20, color='Orange')
ax.set_xlabel('X Label', fontsize=15)
ax.set_ylabel('Y Label', fontsize=15)
ax.tick_params(axis='both', labelsize=7)

plt.show()
```

## What the heck does this mean? `fig.add_subplot(111)`

1. First, `fig.add_subplot()` adds an axes to an existing figure.
2. It's like the add.subplots, but you specify the number of rows, columns and the index (111) means 1 row, 1 column and index 1

```python
fig = plt.figure()

ax = fig.add_subplot(111)

plt.show()
```

# How do I change the tick frequency on the x or y axis?

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticks([-5, -3, -1, 0, 1, 3, 5])
plt.show()
```

## How do you set the y-axis limit?

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_ylim(0, 150)
plt.show()
```

## How do you remove ticks and tick labels?

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

## How do you remove the axis lines, grids, etc?

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_axis_off()
plt.show()
```

## How do you rotate axis text?

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.tick_params(axis='x', rotation=45)
plt.show()
```

## How do you change the scatterplot marker size?

```python
fig, ax = plt.subplots()

ax.scatter(x, y, s=0.1)

plt.show()
```

## How do you color and size every individual point?

```python
fig, ax = plt.subplots()

ax.scatter(x, y, s=0.1, c=y, cmap='coolwarm')

plt.show()
```

## How do you draw horizontal and vertical lines?

```python
fig, ax = plt.subplots()

ax.hlines([-50, 50], min(x), max(x))
ax.vlines([-4], min(x), max(x))
ax.plot(x, y)

plt.show()
```
