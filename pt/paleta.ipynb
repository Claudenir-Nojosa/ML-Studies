```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


def create_big_number_chart(
    percentage,
    title,
    subtitle,
    comparison_text,
    note='',
    color='#5A8EC1'
):
    fig, ax = plt.subplots(figsize=(9, 5))
    ax.axis('off')

    # =========================
    # HEADER
    # =========================
    rect = Rectangle(
        (0.04, 0.88), 0.55, 0.10,
        transform=ax.transAxes,
        color='#9B9B9B',
        alpha=0.9
    )
    ax.add_patch(rect)

    ax.text(
        0.06, 0.93, title,
        fontsize=16,
        color='white',
        ha='left',
        va='center',
        transform=ax.transAxes
    )

    # =========================
    # SUBTITLE
    # =========================
    ax.text(
        0.06, 0.76, subtitle,
        fontsize=12,
        color=color,
        ha='left',
        va='center',
        transform=ax.transAxes
    )

    # =========================
    # BIG NUMBER
    # =========================
    ax.text(
        0.06, 0.56, f'{percentage}%',
        fontsize=110,
        color=color,
        ha='left',
        va='center',
        transform=ax.transAxes,
        fontweight='bold'
    )

    # =========================
    # TEXTO EXPLICATIVO (COM RESPIRO)
    # =========================
    y_position = 0.41

    ax.text(
        0.06, y_position, comparison_text[0],
        fontsize=11,
        color=color,
        ha='left',
        va='center',
        transform=ax.transAxes,
        fontweight='bold'
    )

    y_position -= 0.05
    for line in comparison_text[1:]:
        ax.text(
            0.06, y_position, line,
            fontsize=11,
            color='#6F6F6F',
            ha='left',
            va='center',
            transform=ax.transAxes
        )
        y_position -= 0.04

    # =========================
    # NOTE (SUBIDO)
    # =========================
    if note:
        ax.text(
            0.06, 0.13, note,
            fontsize=8,
            color='#9A9A9A',
            ha='left',
            va='center',
            transform=ax.transAxes,
            style='italic'
        )

    plt.tight_layout()
    return fig, ax


# =========================
# EXEMPLO DE USO
# =========================
fig, ax = create_big_number_chart(
    percentage=68,
    title='Pilot program was a success',
    subtitle='After the pilot program,',
    comparison_text=[
        'of kids expressed interest towards science,',
        'compared to 44% going into the program.'
    ],
    note='Based on survey of 100 students conducted before and after pilot program (100% response rate on both surveys).',
    color='#5A8EC1'
)

plt.show()
```
